```python
import pandas as pd
```

```python
df = pd.read_csv('/content/price_index.csv')
df
```

```python
df = df.drop(["Code", "category_uz", "category_ru", "category_uz_cyrillic"], axis=1)
df
```

```python
import os
from sklearn.linear_model import LinearRegression
import pickle

if 'category_en' not in df.columns:
    raise Exception("❌ 'category_en' column not found!")

os.makedirs('models', exist_ok=True)

for i, row in df.iterrows():
    try:
        category = row['category_en']

        # Get just the time series (skip 'category_en')
        prices = row.drop(labels='category_en')

        # Convert column names (dates) to month index
        prices = prices.dropna()
        prices = prices.reset_index(drop=True)
        prices_df = pd.DataFrame({
            'Month_Num': range(len(prices)),
            'Price': prices.values
        })

        # Train model
        X = prices_df[['Month_Num']]
        y = prices_df['Price']

        model = LinearRegression()
        model.fit(X, y)

        # Save model
        with open(f'models/{category}.pkl', 'wb') as f:
            pickle.dump(model, f)

        print(f"✅ Saved: {category}.pkl")

    except Exception as e:
        print(f"❌ Skipped {row.get('category_en', i)} due to error: {e}")
```

```python
import shutil
shutil.make_archive('all_models', 'zip', 'models')
```
