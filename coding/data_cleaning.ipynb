```python
import pandas as pd
```

```python
df = pd.read_csv('/content/tash_price_index.csv')
df
```

```python
df.rename(columns={
    "Klassifikator": "category_uz",
    "Klassifikator_ru": "category_ru",
    "Klassifikator_en": "category_en",
    "Klassifikator_uzc": "category_uz_cyrillic",
}, inplace=True)

df.head(3)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.columns = [col.replace("М", "M") for col in df.columns]
date_cols = [col for col in df.columns if col.startswith("202")]
new_date_cols = [pd.to_datetime(col + "-01", format="%Y-M%m-%d") for col in date_cols]
col_mapping = dict(zip(date_cols, new_date_cols))
df.rename(columns=col_mapping, inplace=True)
```

```python
df_long = pd.melt(
    df,
    id_vars=['Code', 'category_uz', 'category_ru','category_en', 'category_uz_cyrillic'],
    var_name='date',
    value_name='value'
)

df_long['date'] = pd.to_datetime(df_long['date'])
```

```python
df_long.to_csv("cleaned_data.csv", index=False)
```
